```python
import pandas as pd

# no1
data =pd.read_excel('tugasdataframe.xlsx',sheet_name='data')
data

# memasukan ke csv 
data.to_csv('datasampah1.csv', index=False)
```

```python
# no2
tahun_target = 0
tahun = 2020
for a,b in data.iterrows():
    if b['tahun'] == tahun :
        tahun_target =tahun_target + b['jumlah_produksi_sampah']


datano2=pd.DataFrame([{'tahun':tahun ,'jumlah produksi':tahun_target}])
datano2
# Menyimpan ke CSV dan Excel
datano2.to_excel('datasampah2.xlsx', index=False)
datano2.to_csv('datasampah2.csv', index=False)
```

```python
# no3
jumlah_data = {}

for a, b in data.iterrows():
    tahun = b['tahun']
    jumlah_sampah = b['jumlah_produksi_sampah']

    if tahun not in jumlah_data:
        jumlah_data[tahun] = 0

    jumlah_data[tahun] += jumlah_sampah

datano3 = pd.DataFrame(jumlah_data.items(), columns=['Tahun', 'Jumlah_Sampah'])
datano3

# Menyimpan ke CSV dan Excel
# datano3.to_csv('datasampah3.csv', index=False)
# datano3.to_excel('datasampah3.xlsx', index=False)
```

```python
# no4
data_kota = {}

for a, b in data.iterrows():
    kota          = b['nama_kabupaten_kota']
    tahun         = b['tahun']
    jumlah_sampah = b['jumlah_produksi_sampah']

    if kota not in data_kota:
        data_kota[kota] = {}
        
    if tahun not in data_kota[kota]:
        data_kota[kota][tahun] = 0

    data_kota[kota][tahun] += jumlah_sampah


# memasukan ke csv dan excel
datano4=pd.DataFrame(data_kota.items(),columns=['Nama','Tahun'])
datano4
# datano4.to_csv('datasampah4.csv', index=False)
# datano4.to_excel('datasampah4.xlsx', index=False)
```
